## Load Model

```python
from transformers import pipeline

question_answerer = pipeline("question-answering", "rickyalan/indobert-finetuned-indosquad", framework="tf")
```

## Ask Question

```python
context = """Nama saya Ricky Alan, saya tinggal di Batam, Indonesia.
Saya adalah lulusan prodi Ilmu Komputer dari Universitas Sumatera Utara.
Saya memiliki pemahaman mendalam tentang pemrograman, data analisis dan machine learning.
Saya tertarik untuk menerapkan keahlian saya dan memberikan kontribusi positif di bidang data science."""
```

```python
question = "Dimana saya tinggal?"

question_answerer(context=context, question=question)
```

```python
question = "Pengetahuan apa yang saya miliki?"

question_answerer(context=context, question=question)
```

```python
question = "Saya berkeinginan untuk bekerja di bidang apa?"

question_answerer(context=context, question=question)
```
